```python

tstart = pd.to_datetime('2024/01/25 15', format='%Y/%m/%d %H')
tend = pd.to_datetime('2024/01/26 00', format='%Y/%m/%d %H')
time = pd.date_range(tstart, tend, freq='h')
time=time[:2]
```

```python
t1 = np.zeros((2,2,2))
lon1=[5,6]
lat1=[52.5,53.5]

t2 = np.zeros((2,2,2))
lon2=[5,6]
lat2=[52.5, 53.5]

t2[1,1,1]=1
```

```python
ds1 = xr.Dataset(

    data_vars=dict(

        temperature=(["time", "longitude", "latitude"], t1),


    ),

    coords=dict(
        time=time,
        longitude=lon1,
        latitude=lat1,

    ),

)

ds2 = xr.Dataset(

    data_vars=dict(

        temperature=(["time", "longitude", "latitude"], t2),


    ),

    coords=dict(
        time=time,
        longitude=lon2,
        latitude=lat2,

    ),


)

ds3 = xr.concat((ds1, ds2), dim='latitude')
```

```python
ds3.to_netcdf(filename+'testalert.nc')
```
