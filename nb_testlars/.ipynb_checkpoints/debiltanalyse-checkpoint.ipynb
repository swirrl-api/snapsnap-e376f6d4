```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import matplotlib.dates as mdates
import locale
import requests
import copy
```

```python
filename = '~/work/private/ECMWF_2022_02_28.nc'
```

```python
data = xr.open_dataset(filename)
tstart = pd.to_datetime('2024/01/25 15', format='%Y/%m/%d %H')
tend = pd.to_datetime('2024/01/26 00', format='%Y/%m/%d %H')
time = pd.date_range(tstart, tend, freq='H')
```

```python
plt.rcParams.update({"font.size": 25})
#locale.setlocale(locale.LC_TIME, 'nl_NL.UTF-8')

stations = np.array(['vlissingen', 'hoekvanholland', 'denhelder', 'harlingen', 'delfzijl'])
#outdir2 = '/usr/people/galen/public_html/noordzee/'
outdir = '/usr/people/galen/public_html/noordzee/'
outdir_archive = '/nobackup/users/galen/DCSMV7ECMWFarchief/'  # Extra regel voor opslaan in archief
noens = 24 # number of ens members

for j, stationOI in enumerate(stations):

    datadir = 'https://waterberichtgeving.rws.nl/wbviewer/wb_api.php?request=gettimeseries&lat=51.44&lon=3.6&loc='+ stationOI + '&id=waterlevel_model_delft3dfm_ens&format=csv'

    # =============================================================================
    # Configure station data
    # =============================================================================

    peiltype = np.array(['IP', 'VWP', 'WP', 'AP', 'LAP', 'MHW'])
    peilvlissingen = np.array([290, 310, 330, 370, 410, 530])
    peilhoekvanholland = np.array([180, 200, 220, 280, 365, 510])
    peildenhelder = np.array([150, 170, 190, 260, 345, 450])
    peilharlingen = np.array([220, 240, 270, 330, 390, 490])
    peildelfzijl = np.array([240, 260, 300, 380, 475, 600])

    station = (['vlissingen', 'hoekvanholland', 'denhelder', 'harlingen', 'delfzijl'])
    peilen = np.array([peilvlissingen, peilhoekvanholland, peildenhelder, peilharlingen, peildelfzijl])

    limtabel = xr.Dataset({'peilhoogte': (('station', 'peiltype'), peilen)}, coords={'peiltype': peiltype, 'station': station})

    ip = limtabel.sel(peiltype='IP', station=stationOI).peilhoogte.values
    vwp = limtabel.sel(peiltype='VWP', station=stationOI).peilhoogte.values
    wp = limtabel.sel(peiltype='WP', station=stationOI).peilhoogte.values
    ap = limtabel.sel(peiltype='AP', station=stationOI).peilhoogte.values
    lap = limtabel.sel(peiltype='LAP', station=stationOI).peilhoogte.values
    mhw = limtabel.sel(peiltype='MHW', station=stationOI).peilhoogte.values

    tidefcst = pd.read_csv(datadir, sep=',', names=np.arange(25))
    tidefcst=tidefcst.rename(columns={0: 'time'})
    tidefcst=tidefcst.set_index('time')

    tidefcst=xr.DataArray(tidefcst)

    tidefcst = tidefcst.rename({'dim_1': 'ensmember'})
    tidefcst['time']=pd.to_datetime(tidefcst.time.values)
    tidefcst['time']=pd.to_datetime(tidefcst.time.values)

    tidefcst = tidefcst

    meantide = tidefcst.mean('ensmember')

    timewindow=slice(tidefcst.time[0], tidefcst.time[0]+pd.Timedelta(18, unit='h'))
    meantideOI = meantide.sel(time=timewindow)
    mintidetime = meantideOI.where(meantideOI==meantideOI.min(), drop=True)

    maxtidetimelist = np.empty((18), dtype=type(None))

    for i in range(18):
        tmin = (mintidetime.time+pd.Timedelta(12.4*i, unit='h')).values
        tmax = (mintidetime.time+pd.Timedelta(12.4*(i+1), unit='h')).values
           
        timewindow=slice(tmin[0], tmax[0])
        
        tidefcstOI = tidefcst.sel(time=timewindow)
        maxtideOI = tidefcstOI.max('time')

        maxtideOI = maxtideOI.assign_coords({'nohightide': i+1})
        if i == 0:
            maxtide = maxtideOI
        else: 
            maxtide = xr.concat((maxtide, maxtideOI), dim='nohightide')
        
        meantideOI = meantide.sel(time=timewindow)
        maxtidetime = meantideOI.where(meantideOI==meantideOI.max(), drop=True)
        
        maxtidetimelist[i] = maxtidetime.time.values[0]
        
    maxtidetimelist = pd.to_datetime(maxtidetimelist)
        
    maxtide = maxtide.to_dataset(name='maxtide')
    maxtide = maxtide.assign({'maxtidetime': (('nohightide'), maxtidetimelist)})

    fracip = (xr.where(((maxtide.maxtide>=ip) & (maxtide.maxtide < vwp)), 1, 0)).sum(dim='ensmember')/noens*100
    fracvwp = (xr.where(((maxtide.maxtide>=vwp) & (maxtide.maxtide < wp)), 1, 0)).sum(dim='ensmember')/noens*100
    fracwp = (xr.where(((maxtide.maxtide>=wp) & (maxtide.maxtide < ap)), 1, 0)).sum(dim='ensmember')/noens*100
    fracap = (xr.where(((maxtide.maxtide>=ap) & (maxtide.maxtide < lap)), 1, 0)).sum(dim='ensmember')/noens*100
    fraclap = (xr.where(((maxtide.maxtide>=lap) & (maxtide.maxtide < mhw)), 1, 0)).sum(dim='ensmember')/noens*100
    fracmhw = (xr.where((maxtide.maxtide>=mhw), 1, 0)).sum(dim='ensmember')/noens*100

    # =============================================================================
    # Figure
    # =============================================================================

    latitude = [52, 52, 52, 52, 52]
    longitude=[1, 2, 3, 4, 5]
    fracip = fracip.assign_coords({'station': stationOI, 'latitude': latitude[j], 'longitude': longitude[j]})

    if j == 0:
        fracip_allstn = copy.deepcopy(fracip)
    else:
        fracip_allstn = xr.concat([fracip_allstn, fracip], 'station')

    """
    fig, ax = plt.subplots(1, figsize=(15, 15))
    
    p1 = ax.bar(maxtidetimelist, fracip, width=0.2, bottom=fracvwp+fracwp+fracap+fraclap+fracmhw, color='green', label='IP')
    p2 = ax.bar(maxtidetimelist, fracvwp, width=0.2, bottom=fracwp+fracap+fraclap+fracmhw, color='blue', label='VWP')
    p3 = ax.bar(maxtidetimelist, fracwp, width=0.2, bottom=fracap+fraclap+fracmhw, color='yellow', label='WP')
    p4 = ax.bar(maxtidetimelist, fracap, width=0.2, bottom=fraclap+fracmhw, color='orange', label='AP')
    p5 = ax.bar(maxtidetimelist, fraclap, width=0.2, bottom=fracmhw, color='red', label='LAP')
    p6 = ax.bar(maxtidetimelist, fracmhw, width=0.2, color='black', label='MHW')



    p1 = ax.bar(maxtidetimelist, fracip, width=0.2, color='green', label='IP')
    p2 = ax.bar(maxtidetimelist, fracvwp, width=0.2, bottom=fracip, color='blue', label='VWP')
    p3 = ax.bar(maxtidetimelist, fracwp, width=0.2, bottom=fracip+fracvwp, color='yellow', label='WP')
    p4 = ax.bar(maxtidetimelist, fracap, width=0.2, bottom=fracip+fracvwp+fracwp, color='orange', label='AP')
    p5 = ax.bar(maxtidetimelist, fraclap, width=0.2, bottom=fracip+fracvwp+fracwp+fracap, color='red', label='LAP')
    p6 = ax.bar(maxtidetimelist, fracmhw, width=0.2, bottom=fracip+fracvwp+fracwp+fracap+fraclap, color='black', label='MHW')



    plt.legend()

    ax.set_xticks(maxtidetimelist+pd.Timedelta(1, unit='h'))  # convert to local time
    plt.xticks(rotation=90)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%a-%d %h:%M'))

    ax.set_ylim([0, 100])

    ax.set_xlabel('Tijd hoogwater (LT)')
    ax.set_ylabel('Waterstandskans (%)')

    plt.axhline(25, linestyle='--', color='black')
    plt.axhline(20, linestyle='-.', color='black')

    plt.title('Overschrijdingskans '+ stationOI + ' op basis van \n DCSMV7-ECMWF \n run ' + pd.to_datetime(tidefcst.time.values[0]).strftime('%d-%m-%Y %h') + 'UTC')

    plt.tight_layout()

    plt.savefig(outdir + stationOI + '.png')
    uil
    #timestamp = pd.to_datetime(tidefcst.time.values[0]).strftime('%Y_%m_%d_%H')
    
    #plt.savefig(outdir_archive + timestamp + '_' + stationOI + '.png')  # Deze regel kan weg als ik overga op nieuw archief

    #  plt.savefig(outdir + timestamp + '_' + stationOI + '.png')   # Extra regel voor opslaan in archief
    """
```
